# Stack and unstacking using Groupby

### Plot the number of gold medals won by the US male and female atheletes througout the history of Olympics

```python
oo[(oo.NOC == 'USA') & (oo.Medal == 'Gold')].groupby(['Edition','Gender']).size().unstack('Gender', fill_value=0).plot(kind='bar');
```

### Plot the 5 atheletes who have won the most gold medals over the history of Olympics. When there is a tie, consider the number of silver medals and then bronze medals.

```python
g = oo.groupby(['Athlete','Medal']).size().unstack('Medal', fill_value=0)
g.head()
```

```python
g.sort_values(['Gold','Silver','Bronze'], ascending=False)[['Gold','Silver','Bronze']].head(5).plot(kind='bar');
```

# In every Olympics, which US athlete has won the most total number of medals? Include the athlete's discipline in solution

```python
usa = oo[oo.NOC == 'USA']
usa.head()
```

```python
usa = usa.groupby(['Edition','Athlete','Medal']).size().unstack('Medal', fill_value=0)
usa['Total'] = usa['Gold'] + usa['Silver'] + usa['Bronze']
```

```python
usa.head()
```

```python
usa.reset_index(inplace=True)
```

```python
for year, group in usa.groupby('Edition'):
    print(group.sort_values('Total', ascending=False)[:1])
```

```python
type(usa.groupby('Edition'))
```

```python
for year, group in usa.groupby('Edition'):
    print(group.sort_values('Total', ascending=False).head(1))
```

```python
# Store this in a list fashion using list comprehension

final_list = [group.sort_values('Total', ascending=False).head(1) for year, group in usa.groupby('Edition')]
```

```python
# Once you have the list, then you can populate the dataframe from that list element using the below logic.

final_df = pd.DataFrame()
for i in final_list:
    final_df = final_df.append(i)
```
